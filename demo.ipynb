```python
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.z(1)
qc.cx(1, 0)
qc.h(1)
```

```python
qc.measure(0, 0)
```

```python
# Run the circuit on AerSimulator and print the single-shot measured bit
backend = AerSimulator()
tqc = transpile(qc, backend)
job = backend.run(tqc, shots=1)
result = job.result()
counts = result.get_counts()
bit = next(iter(counts.keys()))
print(bit)
```
